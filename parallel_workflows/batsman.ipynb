```python
from langgraph.graph import StateGraph,START,END
from typing import TypedDict, List
```

```python
class BatsmanState(TypedDict):
    name: str
    runs: int
    balls_faced: int
    fours:int
    sixes:int

    sr:float
    bpb:float
    boundarypercentage:float
```

```python
def calc_sr(state:BatsmanState)->BatsmanState:
    state['sr']=(state['runs']/state['balls_faced'])*100
    return {'sr':state['sr']}
```

```python
def calc_bpb(state:BatsmanState)->BatsmanState:
    state['bpb']=state['balls_faced']/(state['fours']+state['sixes'])
    return {'bpb':state['bpb']}
```

```python
def calc_boundarypercentage(state:BatsmanState)->BatsmanState:
    state['boundarypercentage']=(state['fours']+state['sixes'])/state['balls_faced']*100
    return {'boundarypercentage':state['boundarypercentage']}
```

```python
def summary(state:BatsmanState)->BatsmanState:
    print(f"Name: {state['name']}")
    print(f"Runs: {state['runs']}")
    print(f"Balls Faced: {state['balls_faced']}")
    print(f"Fours: {state['fours']}")
    print(f"Sixes: {state['sixes']}")
    print(f"Strike Rate: {state['sr']}")
    print(f"Balls per Boundary: {state['bpb']}")
    print(f"Boundary Percentage: {state['boundarypercentage']}")
    return {}
```

```python
graph=StateGraph(BatsmanState)

graph.add_node('calc_sr',calc_sr)
graph.add_node('calc_bpb',calc_bpb)
graph.add_node('calc_boundarypercentage',calc_boundarypercentage)
graph.add_node('summary',summary)

graph.add_edge(START,'calc_bpb')
graph.add_edge(START,'calc_sr')
graph.add_edge(START,'calc_boundarypercentage') 

graph.add_edge('calc_bpb','summary')
graph.add_edge('calc_sr','summary') 
graph.add_edge('calc_boundarypercentage','summary') 
graph.add_edge('summary',END)

workflow=graph.compile()
```

```python
initial_state={
    'name': 'Virat Kohli',
    'runs': 100,
    'balls_faced': 60,  
    'fours': 10,
    'sixes': 5,
}
```

```python
workflow.invoke(initial_state)
```
